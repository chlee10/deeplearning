```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('e:/data/final/labels.csv')
df.head()
```

```python
df.isnull().values.any()
```

```python
df.breed.nunique()
```

```python
y=pd.value_counts(df['breed'])

fig, ax = plt.subplots()
fig.set_size_inches(15, 9)
sns.set_style("whitegrid")

ax = sns.barplot(x = y.index, y = y, data = df)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90, fontsize = 8)
ax.set(xlabel='Dog Breed', ylabel='Count')
ax.set_title('Distribution of Dog breeds')

top_breeds = sorted(list(df['breed'].value_counts().head(16).index))
print(top_breeds)
```

```python
import matplotlib.pylab as plt

plt.rcParams["figure.figsize"] = (15,9)
plt.rcParams['axes.grid'] = False
a=df['breed'].value_counts().to_frame()
a.columns=['all']

SEED=123

x_train, x_validation, y_train, y_validation = train_test_split(df, df['breed'], test_size=0.2, random_state=SEED)

#calculate the value counts for train and validation data
a['train']=y_train.value_counts().sort_index()

a['validation'] = y_validation.value_counts().sort_index()

a[['train','validation']].plot(kind='bar',stacked=True)

plt.show()
```
